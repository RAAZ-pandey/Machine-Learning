```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

```python
customer_data = pd.read_csv('agglomarative.csv')
```

```python
# Preview shape and contents
print("Shape:", customer_data.shape)
```

```python
print(customer_data.head())
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendrogram")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

from sklearn.cluster import AgglomerativeClustering
import numpy as np

# Ensure data is properly formatted
data = np.array(data)  # Convert to numpy array if not already
if len(data.shape) == 1:
    data = data.reshape(-1, 1)  # Reshape if 1D
    
# Assuming df is your original DataFrame
data = customer_data.select_dtypes(include=[float, int])  # or manually select features

# Agglomerative Clustering
cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
labels_ = cluster.fit_predict(data)
```

```python
labels_
```

```python
# Plot clustering result
plt.figure(figsize=(10, 7))
plt.scatter(data[:, 0], data[:, 1], c=labels_, cmap='rainbow')
plt.title("Agglomerative Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
