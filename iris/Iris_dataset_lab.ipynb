```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#loading dataset
df=pd.read_csv("Iris.csv")
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.hist(figsize=(10, 6), bins=20, edgecolor='black')
plt.suptitle('Feature Distributions')
plt.subplots_adjust(bottom=0.1)
plt.show()
```

```python
# Line Chart of Sepal Length
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['SepalLengthCm'],linestyle='-', color='b')
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.title("Line Chart of Sepal Length")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(7, 4))
sns.boxplot(data=df, orient='h')
plt.title("Boxplot of Features")
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue='Species', diag_kind='kde')
plt.show()
```

```python
#Heatmap
plt.figure(figsize=(8, 6))
numeric_df=df.select_dtypes(include=['float64','int64'])
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```
